```python
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some sample data
np.random.seed(42)
data1 = np.random.normal(loc=0.0, scale=1.0, size=(300, 1))  # Cluster 1
data2 = np.random.normal(loc=5.0, scale=1.0, size=(300, 1))  # Cluster 2
data3 = np.random.normal(loc=10.0, scale=1.0, size=(300, 1))  # Cluster 3
data = np.vstack((data1, data2, data3))
```

```python
# Fit GMM
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(data)

# Predict cluster labels
labels = gmm.predict(data)
```

```python
# Plot the data with cluster labels
plt.scatter(data, np.zeros_like(data), c=labels, cmap='viridis', s=10)
plt.title('Gaussian Mixture Model Clustering')
plt.xlabel('Data Value')
plt.ylabel('Cluster Label')
plt.show()
```

```python
# Print cluster means
print("Cluster Means:")
print(gmm.means_)
```
